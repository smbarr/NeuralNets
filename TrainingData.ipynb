Training data are circles with noise

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tagPoints(x,y):
    N = x.shape[0]
    r1 = np.full(N, 1)
    r2 = np.full(N, 1)
    noise = 0.1*np.random.randn(N)
    tag1 = np.where(np.sqrt((x-0.5)**2+(y-0.75)**2)-r1+noise > 0, 0, 1)
    tag2 = np.where(np.sqrt((x+0.5)**2+(y+0.75)**2)-r2+noise > 0, 0, 1)
    tag = np.where((tag1 > 0) | (tag2 > 0), 1, 0)
    return tag
```

```python
x = 5.0*(np.random.uniform(size=5000)-0.5)
y = 5.0*(np.random.uniform(size=5000)-0.5)
tag = tagPoints(x,y)
```

```python
plt.figure(figsize=(10,10))
plt.gca().set_aspect('equal')
plt.scatter(x, y, c=tag, cmap=plt.cm.RdYlBu)
```

```python
np.savetxt("train_data.dat", np.hstack([x[:,None],y[:,None],tag[:,None]]))
```
