```python
import numpy as np
from tqdm import trange
import matplotlib.pyplot as plt
```

```python
def sigmoid(x, deriv=False):
    val = None
    if not deriv:
        val = 1.0/(1.0+np.exp(-x))
    else:
        val = np.exp(-x)/((np.exp(-x)+1.0)**2)
    return val
```

```python
def MSE(y, target, deriv=False):
    n = y.shape[0]
    val = None
    if not deriv:
        val = np.sum((target-y)**2)/n
    else:
        val = -2.0*(target-y)/n
    return val
```

```python
class NeuralNet:
    def __init__(self, layers, params):
        self.params = params
        self.nLayers = len(layers)
        self.W = [np.random.randn(layers[i],layers[i+1]) for i in range(self.nLayers-1)]
        self.b = [np.random.randn(layers[i+1]) for i in range(self.nLayers-1)]
        self.z = [None for n in range(self.nLayers-1)]
        self.a = [None for n in range(self.nLayers-1)]
        
    def eval(self,x):
        self.z[0] = x @ self.W[0] + self.b[0]
        self.a[0] = self.params["activation"](self.z[0])
        for n in range(1,self.nLayers-1):
            self.z[n] = self.a[n-1] @ self.W[n] + self.b[n]
            self.a[n] = self.params["activation"](self.z[n])
        return self.a[-1]
    
    def loss(self, x, target):
        y = self.eval(x)
        return(self.params["loss"](y, target))
    
    def backProp(self, x, target):
        y = self.eval(x)
        dl_dw = [None for n in range(self.nLayers-1)]
        dl_dw[self.nLayers-2] = self.params["loss"](y, target, True) * self.params["activation"](self.z[self.nLayers-2], True)
        for n in range(self.nLayers-2,0,-1):
            if len(dl_dw[n].shape) == 2:
                dl_dw[n-1] = (self.W[n].T * dl_dw[n][:,None]) * self.params["activation"](self.z[n-1], True)[:,None]
            else:
                dl_dw[n-1] = (self.W[n].T * dl_dw[n]) * self.params["activation"](self.z[n-1], True)[:,None]
        
        self.W[0] = self.W[0] - self.params["learningRate"]*(x.T @ dl_dw[0].sum(axis=2))
        self.b[0] = self.b[0] - dl_dw[0].sum(axis=2)
        for n in range(1,self.nLayers-1):
            if len(dl_dw[n].shape) > 2:
                self.W[n] = self.W[n] - self.params["learningRate"]*(self.a[n-1].T @ dl_dw[n].sum(axis=2))
                self.b[n] = self.b[n] - dl_dw[n].sum(axis=2)
            else:
                self.W[n] = self.W[n] - self.params["learningRate"]*(self.a[n-1].T @ dl_dw[n])
                self.b[n] = self.b[n] - dl_dw[n]
```

```python
net = NeuralNet(
    [2,300,1], 
    {
        "activation": sigmoid,
        "loss": MSE,
        "learningRate": 0.1
    }
)
```

```python
data = np.loadtxt("train_data.dat")
x = data[:,:2]
y = data[:,2][None,:].T
```

```python
loss = []
with trange(1000) as pbar:
    for i in pbar:
        net.backProp(x, y)
        loss.append(net.loss(x,y))
        pbar.set_description("Loss: %f"%(loss[-1]))
pred = net.eval(x)
```

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 16
plt.figure(figsize=(10,10/((np.sqrt(5)+1)/2)))
plt.title("Training Loss")
plt.plot(loss, 'k')
plt.show()
```
